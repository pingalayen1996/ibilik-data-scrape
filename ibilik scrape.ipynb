```python
#import libraries
import bs4
import pandas as pd
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import numpy as np
```

```python
#create empty lists
containers = []
listing_id = []
title = []
link = []
publish_date = []
location = []
customer_preferences = []
room_type = []
room_perks = []
room_price = []
url2 = []
```

```python
#number of pages on the website to scrape through.
range = np.arange(1,551)
```

```python
#specifies link and state to scrape through. Change "selangor" to "kuala_lumpur" or any state you'd like to scrape through
for i in range:
   url2.append("https://ibilik.my/rooms/selangor?page=" + str(i))
```

```python
#scraper doing its thang
for i in url2:
#     my_url = i
#     my_url = "https://ibilik.my/rooms/kuala_lumpur?page=" + "1"
    uClient = uReq(i)
    page_html = uClient.read()
    uClient.close()
    page_soup = soup(page_html,"html.parser")
    containers = page_soup.findAll("div",{"class":"home-list-pop"})
    for contain in containers:
#         try:
        listing_id.append(contain.findAll("div",{"class":"home-list-pop-desc inn-list-pop-desc"})[0].a["data-content-id"])
        title.append(contain.findAll("div",{"class":"home-list-pop-desc inn-list-pop-desc"})[0].h3.text)
        link.append(contain.findAll("div",{"class":"home-list-pop-desc inn-list-pop-desc"})[0].a["href"])
        publish_date.append(contain.findAll("div",{"class":"home-list-pop-desc inn-list-pop-desc"})[0].i.text)
        location.append(contain.findAll("p",{"class":""})[0].text)
        customer_preferences.append(contain.findAll("p",{"class":""})[1].text)
        room_price.append(contain.findAll("div",{"class":"room_price"})[0].span.text)
        try:
            room_type.append(contain.findAll("p",{"class":""})[2].text)
        except IndexError:
            room_type.append("0")
        try:
            room_perks.append(contain.findAll("p",{"class":""})[3].text)
        except IndexError:
            room_perks.append("0")
#         room_perks.append(contain.findAll("p",{"class":""})[3].text)
        
#             break
#         except:
#             listing_id.append(contain.findAll("div",{"class":"home-list-pop-desc inn-list-pop-desc"})[0].a["data-content-id"])
#             title.append(contain.findAll("div",{"class":"home-list-pop-desc inn-list-pop-desc"})[0].h3.text)
#             link.append(contain.findAll("div",{"class":"home-list-pop-desc inn-list-pop-desc"})[0].a["href"])
#             publish_date.append(contain.findAll("div",{"class":"home-list-pop-desc inn-list-pop-desc"})[0].i.text)
#             location.append(contain.findAll("p",{"class":""})[0].text)
#             customer_preferences.append(contain.findAll("p",{"class":""})[1].text)
#             room_type.append(contain.findAll("p",{"class":""})[2].text)
#             room_perks.append(np.nan)
#             room_price.append(contain.findAll("div",{"class":"room_price"})[0].span.text)
```

```python
#create dataframe
df_data = pd.DataFrame(
    {'listing_id': listing_id,
     'title': title,
     'link': link,
     'publish_date': publish_date,
     'location': location,
     'preferences': customer_preferences,
     'room_type': room_type,
     'room_perks': room_perks,
     'room_price': room_price,
    })
```

```python
df_data.head()
```

```python
#functions to help with data cleaning

def third_location(x):
    try:
        x.split(",")[2]
    except IndexError:
        ""

def striplist(l):
    return([x.strip() for x in l])
```

```python
#data cleaning
df_data['publish_date'] = df_data['publish_date'].apply(lambda x: x[14:])
df_data['location'] = df_data['location'].apply(lambda x: x.strip())
df_data['preferences'] = df_data['preferences'].apply(lambda x: x.strip())
df_data['room_type'] = df_data['room_type'].apply(lambda x: x.strip())
df_data['room_perks'] = df_data['room_perks'].apply(lambda x: x.strip())
```

```python
#data cleaning - re-indexes column values for some rows which were indexed wrongly.
rows_to_shift = df_data[df_data['room_perks']=='0'].index
df_data.iloc[rows_to_shift,5:8] = df_data.iloc[rows_to_shift,5:8].shift(periods=1,axis=1)
```

```python
#creating new fields - splitting address into 3 lines
df_data['location_1'] = df_data['location'].apply(lambda x: x.split(",")[0])
df_data['location_2'] = df_data['location'].apply(lambda x: x.split(",")[1])
df_data['location_3'] = df_data['location'].apply(lambda x: third_location(x))
```

```python
#data cleaning - converts preferences column into list format for easier extraction
df_data['room_perks_list'] = df_data['room_perks'].apply(lambda x: str(x).strip().lower().split(","))
df_data['preferences_list'] = df_data['preferences'].apply(lambda x: str(x).strip().lower().split(","))
```

```python
#creating new fields - True/False if value of column name is in list of preferences.
df_data['indian'] = df_data.preferences_list.apply(lambda x: striplist(x)).map(set(['indian']).issubset)
df_data['malay'] = df_data.preferences_list.apply(lambda x: striplist(x)).map(set(['malay']).issubset)
df_data['chinese'] = df_data.preferences_list.apply(lambda x: striplist(x)).map(set(['chinese']).issubset)
df_data['malaysian'] = df_data.preferences_list.apply(lambda x: striplist(x)).map(set(['malaysian']).issubset)
df_data['student'] = df_data.preferences_list.apply(lambda x: striplist(x)).map(set(['student']).issubset)
df_data['other'] = df_data.preferences_list.apply(lambda x: striplist(x)).map(set(['other']).issubset)
df_data['female'] = df_data.preferences_list.apply(lambda x: striplist(x)).map(set(['single female']).issubset)
df_data['male'] = df_data.preferences_list.apply(lambda x: striplist(x)).map(set(['single male']).issubset)
df_data['non-malaysian'] = df_data.preferences_list.apply(lambda x: striplist(x)).map(set(['non-malaysian']).issubset)
```

```python
#dropping columns
df_data = df_data.drop(['preferences_list','room_perks_list'],axis=1)
```
